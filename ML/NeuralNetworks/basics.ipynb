## Neural Network Basics
### Multilayer Perceptron (MLP)
* Neuron : it is the main building block of a neural network, which receives input, processes it, and produces a single output, it can act as logic gates like AND, OR, NOT.

* Perceptron : it is a type of artificial neuron that take multiple numeric inputs such that each input has its own weight, the perceptron computes a weighted sum (z) of the inputs and applies an activation function to determine the output. ```
z = w1*x1 + w2*x2 + ... + wn*xn + b where w1, w2, ..., wn are the weights, x1, x2, ..., xn are the inputs, and b is the bias.```

* Bias : it is a constant value added to the weighted sum (z) before applying the activation function. It helps the perceptron to shift the activation function, allowing it to better fit the data.

* TLU (Threshold Logic Unit) : it is a type of perceptron that uses a step activation function. If the weighted sum (z) is greater than or equal to a certain threshold, the output is 1; otherwise, it is 0. ``` y = 1 if (Wx + b) >= threshold else 0 ```

* The perceptron Output is determined by applying an activation function to the weighted sum (z). Common activation functions include the step function, sigmoid, ReLU, and others. ``` y = activation_function(Wx + b) ```

* Perceptrons are stacked together to form a multilayer perceptron (MLP), which consists of multiple layers of neurons. Each layer processes the outputs from the previous layer, allowing the network to learn complex patterns in the data.

* Layers : Neurons are organized into layers to form a multilayer perceptron (MLP). An MLP typically consists of:
  - Input Layer: Receives the initial data.
  - Hidden Layers: Intermediate layers that process inputs from the previous layer.
  - Output Layer: Produces the final output of the network.



### Learning Process

* Forward Propagation : In this phase, the input data is passed through the network layer by layer. Each neuron computes its output based on the weighted sum of its inputs and the activation function. The final output is produced at the output layer.

* Loss Funnction : After the final output is produced, the network compares it to the actual target value, the loss function is the difference between the predicted output and the actual target value. 

* Backpropagation : In this phase, the network adjusts its weights and biases based on the error calculated by the loss function. The error is propagated backward through the network, and the weights are updated using optimization algorithms like gradient descent to minimize the loss.

* z=wx+b, y_pred = activation_function(z), loss = loss_function(y_pred, y_true), then the loss is minimized by updating weights and biases using backpropagation (differentiating Loss w.r.t. weights) : ``` dL/dw​=dL/dy_pred * dy_pred/dz * dz/dw ```

* Then the weights are updated as: ``` W_new = W_old - learning_rate * (y_pred - y_true) * x  = W_old - learning_rate(dL/dW) ```

* Activation Function : It is a non-linear mathematical function applied to the weighted sum (z) of inputs in a neuron. It introduces non-linearity into the network, allowing it to learn complex patterns. it is required as without it, the network collapses into a single-layer perceptron, limiting its ability to model complex relationships.
